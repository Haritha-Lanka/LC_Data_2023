```python
import pandas as pd

df = pd.read_csv('../resources/popular_board_game.csv')

num_rows = len(df.index) #variable "num_rows" = the number of rows in the dataframe
low_information_cols = [] #empty list

for col in df.columns: #loop over the columns in the dataset
    print(f"STARTING COLUMN: {col}")
    cnts = df[col].value_counts(dropna=False) #"cnts" = report all unique values that appear in the column 
                                                #and how many times each one appears
    print(f"cnts = {cnts}")
    print("---")
    
    print(f"cnts/num_rows = {cnts/num_rows}") #"cnts/num_rows" converts each number from the "cnts" variable 
                                                #as the percentage of the column represented by that value
    print("---")
    
    print(f"(cnts/num_rows).iloc[0] = {(cnts/num_rows).iloc[0]}") #cnts/num_rows).iloc[0] gets the number in the first row
                                                                    #of cnts/num_rows. Since .value_counts() sorts values
                                                                    #in descending order, we know that this value is the
                                                                    #largest percentage of the column
    print("---")
    
    top_pct = (cnts/num_rows).iloc[0] #assign the value of (cnts/num_rows).iloc[0] to the variable "top_pct"
    print(f"top_pct = {top_pct}")
    print("---")
    
    #NOTE: I changed the top_pct threshold value from .95 to .3 because the dataset I'm using didn't have 
        # any values that took up >95% of any column. This allows me to see what happens when this conditional
        # is triggered. In production, I would need to put this back to .95
        
    if top_pct > 0.30: #if the most common value in the column represents more than 95% of all rows
        low_information_cols.append(col) #append the name of the column to the list "low_information_cols"
        print("Low data column:")
        print("value with greatest representation:")
        print('{0}: {1:.5f}%'.format(col, top_pct*100))#print the name of the column and what percentage 
                                                        #the top value represents
        print("---")
        print("CNTS:")
        print(cnts)
        print("---")
        print(f"{col} complete")
        print("---")
```
